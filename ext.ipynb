```python
from PyPDF2 import PdfReader



# Load the PDF file

reader = PdfReader("Datasetlegal.pdf")



# Extract text from all pages

pdf_text = ""

for page in reader.pages:

    pdf_text += page.extract_text()
```

```python
import re
import pandas as pd

# Define patterns for sections, definitions, and illustrations
section_pattern = r"^\d+\.\s.*"  # Sections often start with numbers followed by a dot and title

# Split content by newlines to analyze line by line
lines = pdf_text.split("\n")

# Variables to store parsed data
structured_data = []
current_section = ""

# Iterate through lines to categorize content
for line in lines:
    line = line.strip()  # Clean up whitespace
    if re.match(section_pattern, line):
        # Store the previous section if it exists
        if current_section:
            structured_data.append({
                "Section": current_section
            })
        # Update current section
        current_section = line
# Append the last section if it exists
if current_section:
    structured_data.append({
        "Section": current_section
    })

# Convert structured data to a DataFrame for better visualization and manipulation
df = pd.DataFrame(structured_data)

# Display a preview of the DataFrame
df.head()
```

```python
def extract_illustrations_from_pdf(txt):
    # Extract text from PDF
    text = txt

    # Split the text into sections
    sections = text.split("Illustrations.")
    results = []

    for section in sections:
        # Extract the section heading and first illustration
        lines = section.splitlines()
        heading = lines[0].strip() if lines else "Unknown Section"
        illustration = next((line.strip() for line in lines if line.strip().startswith("(")), "No illustration found")
        results.append((heading, illustration))

    return results

# Extract illustrations
illustrations = extract_illustrations_from_pdf(pdf_text)

```

```python
df1 = pd.DataFrame(illustrations)
```
